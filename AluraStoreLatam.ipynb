### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# ingresos totales de cada tienda
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

df_ingresos = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"])
print(df_ingresos)

df_ingresos.plot(kind='bar', x='Tienda', y='Ingreso Total', legend=False, color='skyblue')
```

# 2. Ventas por categoría

```python
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

df_ingresos = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"])
print(df_ingresos)

df_ingresos.plot(kind='bar', x='Tienda', y='Ingreso Total', legend=False, color='skyblue')

ventas_categoria_t1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_t4 = tienda4['Categoría del Producto'].value_counts()

print(" Categoría más popular por tienda:")
print("Tienda 1:", ventas_categoria_t1.idxmax(), "-", ventas_categoria_t1.max(), "ventas")
print("Tienda 2:", ventas_categoria_t2.idxmax(), "-", ventas_categoria_t2.max(), "ventas")
print("Tienda 3:", ventas_categoria_t3.idxmax(), "-", ventas_categoria_t3.max(), "ventas")
print("Tienda 4:", ventas_categoria_t4.idxmax(), "-", ventas_categoria_t4.max(), "ventas")

import pandas as pd

import matplotlib.pyplot as plt

categorias_populares = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Categoría más popular": [
        ventas_categoria_t1.idxmax(),
        ventas_categoria_t2.idxmax(),
        ventas_categoria_t3.idxmax(),
        ventas_categoria_t4.idxmax()
    ],
    "Ventas": [
        ventas_categoria_t1.max(),
        ventas_categoria_t2.max(),
        ventas_categoria_t3.max(),
        ventas_categoria_t4.max()
    ]
})
etiquetas = [
    f"{cat} ({tienda})"
    for cat, tienda in zip(categorias_populares["Categoría más popular"], categorias_populares["Tienda"])
]

plt.figure(figsize=(8, 6))
plt.pie(categorias_populares["Ventas"],
        labels=etiquetas,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors)
plt.title("Distribución de las Categorías más Populares por Tienda")
plt.axis('equal')
plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_t1 = tienda["Producto"].value_counts()
ventas_t2 = tienda2["Producto"].value_counts()
ventas_t3 = tienda3["Producto"].value_counts()
ventas_t4 = tienda4["Producto"].value_counts()

print("Productos más y menos vendidos:")

print("\nTienda 1:")
print("Más vendido:", ventas_t1.idxmax(), "-", ventas_t1.max(), "ventas")
print("Menos vendido:", ventas_t1.idxmin(), "-", ventas_t1.min(), "ventas")

print("\nTienda 2:")
print("Más vendido:", ventas_t2.idxmax(), "-", ventas_t2.max(), "ventas")
print("Menos vendido:", ventas_t2.idxmin(), "-", ventas_t2.min(), "ventas")

print("\nTienda 3:")
print("Más vendido:", ventas_t3.idxmax(), "-", ventas_t3.max(), "ventas")
print("Menos vendido:", ventas_t3.idxmin(), "-", ventas_t3.min(), "ventas")

print("\nTienda 4:")
print("Más vendido:", ventas_t4.idxmax(), "-", ventas_t4.max(), "ventas")
print("Menos vendido:", ventas_t4.idxmin(), "-", ventas_t4.min(), "ventas")

import matplotlib.pyplot as plt

def graficar_top5(ventas, titulo):
    top5 = ventas.head(5)
    plt.figure(figsize=(8, 4))
    top5.sort_values().plot(kind='barh', color='lightgreen')
    plt.title(f"Top 5 productos más vendidos - {titulo}")
    plt.xlabel("Cantidad de ventas")
    plt.tight_layout()
    plt.show()

# Graficar para cada tienda
graficar_top5(ventas_t1, "Tienda 1")
graficar_top5(ventas_t2, "Tienda 2")
graficar_top5(ventas_t3, "Tienda 3")
graficar_top5(ventas_t4, "Tienda 4")


def graficar_bottom5(ventas, titulo):
    bottom5 = ventas.sort_values().head(5)  # ordenar de menor a mayor y tomar los 5 primeros
    plt.figure(figsize=(8, 4))
    bottom5.plot(kind='barh', color='salmon')
    plt.title(f"Bottom 5 productos menos vendidos - {titulo}")
    plt.xlabel("Cantidad de ventas")
    plt.tight_layout()
    plt.show()

graficar_bottom5(ventas_t1, "Tienda 1")
graficar_bottom5(ventas_t2, "Tienda 2")
graficar_bottom5(ventas_t3, "Tienda 3")
graficar_bottom5(ventas_t4, "Tienda 4")
```
